```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
harvey_data = pd.read_csv("Resources/harvey.csv",index_col=1)
harvey_data
```

```python
fig, ax = plt.subplots()

ax.bar(harvey_data.index, harvey_data['unemployment'])

ax.set_title ("Unemployment in Harris County", fontsize = 18)

ax.set_ylabel("Percentage of Unemployed")

plt.show
```
