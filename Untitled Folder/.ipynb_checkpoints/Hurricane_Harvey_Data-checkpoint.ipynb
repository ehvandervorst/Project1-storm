```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests 
from census import Census

# Census API Key
from census_key import census_key
c1 = Census(census_key, year=2014)
```

```python
census_data = c1.acs5.get(("NAME", "B25035_001E", 
                          "B25058_001E", "B25064_001E",
                         "B25077_001E", "B25088_002E"), {'for': 'county:*'})
census_2014_harvey = pd.DataFrame(census_data)

census_2014_harvey = census_2014_harvey.rename(columns={"B25035_001E": "Median House Construction",
                                                      "B25058_001E": "Median Contract Rent",
                                                      "B25064_001E": "Median Gross Rent",
                                                      "B25077_001E": "Median Home Value",
                                                      "B25088_002E": "Median Monthly Owner Cost",
                                                      "NAME": "Name", "county": "county"})
census_2014_harvey.head()
```

```python
# Census API Key
c2 = Census(census_key, year=2017)
```

```python
census_data = c2.acs5.get(("NAME", "B25035_001E", 
                          "B25058_001E", "B25064_001E",
                         "B25077_001E", "B25088_002E"), {'for': 'county:*'})
census_2017_harvey = pd.DataFrame(census_data)

census_2017_harvey = census_2017_harvey.rename(columns={"B25035_001E": "Median House Construction",
                                                      "B25058_001E": "Median Contract Rent",
                                                      "B25064_001E": "Median Gross Rent",
                                                      "B25077_001E": "Median Home Value",
                                                      "B25088_002E": "Median Monthly Owner Cost",
                                                      "NAME": "Name", "county": "county"})
census_2017_harvey.head()
```

```python
# Census API Key
c3 = Census(census_key, year=2018)
```

```python
census_data = c3.acs5.get(("NAME", "B25035_001E", 
                          "B25058_001E", "B25064_001E",
                         "B25077_001E", "B25088_002E"), {'for': 'county:*'})
census_2018_harvey = pd.DataFrame(census_data)

census_2018_harvey = census_2018_harvey.rename(columns={"B25035_001E": "Median House Construction",
                                                      "B25058_001E": "Median Contract Rent",
                                                      "B25064_001E": "Median Gross Rent",
                                                      "B25077_001E": "Median Home Value",
                                                      "B25088_002E": "Median Monthly Owner Cost",
                                                      "NAME": "Name", "county": "county"})
census_2018_harvey.head()
```

```python
merge_one_data = pd.merge(census_2014_harvey, census_2017_harvey, how="right", on="Name")

merge_one_data.head()
```

```python
harvey_complete_housing = pd.merge(merge_one_data, census_2018_harvey, how="right", on="Name")

harvey_complete_housing.head()
```

```python
harvey_reduced = harvey_complete_housing[["Name", "Median House Construction_x", "Median Contract Rent_x", "Median Home Value_x",
                                        "Median House Construction_y", "Median Contract Rent_y", "Median Home Value_y",
                                        "Median House Construction", "Median Contract Rent", "Median Home Value"]]
harvey_reduced.head()  
```

```python
harvey_renamed = harvey_reduced.rename(columns={"Median House Construction_x": "Construction_2010",
                                             "Median Contract Rent_x": "Rent_2010",
                                             "Median Home Value_x": "Value_2010",
                                             "Median House Construction_y": "Construction_2012",
                                             "Median Contract Rent_y": "Rent_2012",
                                             "Median Home Value_y": "Value_2012",
                                             "Median House Construction": "Construction_2015",
                                             "Median Contract Rent": "Rent_2015",
                                             "Median Home Value": "Value_2015"})
harvey_renamed.head()
```

```python
harvey_clean = harvey_renamed.loc[harvey_renamed["Name"] == "Harris County, Texas"]
harvey_clean
```

```python
harvey_housing_df = pd.DataFrame({"Years": [2014, 2017, 2018],
                        "Median Construction": [1982, 1983, 1984],
                        "Median Contract Rent": [738, 827, 875],
                        "Median Home Value": [133400, 154100, 165300]})
harvey_housing_df
```

```python
years = ["2014", "2017", "2018"]
construction = [1982, 1983, 1984]
x_axis = np.arange(len(construction))

plt.bar(x_axis, construction, color="b", align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, years)


plt.title("Harris County Median House Construction")
plt.xlabel("Years")
plt.ylabel("Costs")

plt.savefig('harvey construction')
```
