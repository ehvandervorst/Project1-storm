```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests 
from census import Census

# Census API Key
from census_key import census_key
c1 = Census(census_key, year=2010)
```

```python
census_data = c1.acs5.get(("NAME", "B25035_001E", 
                          "B25058_001E", "B25064_001E",
                         "B25077_001E", "B25088_002E"), {'for': 'county:*'})
census_2010_sandy = pd.DataFrame(census_data)

census_2010_sandy = census_2010_sandy.rename(columns={"B25035_001E": "Median House Construction",
                                                      "B25058_001E": "Median Contract Rent",
                                                      "B25064_001E": "Median Gross Rent",
                                                      "B25077_001E": "Median Home Value",
                                                      "B25088_002E": "Median Monthly Owner Cost",
                                                      "NAME": "Name", "county": "county"})
census_2010_sandy.head()
```

```python
# Census API Key
c2 = Census(census_key, year=2012)
```

```python
census_data = c2.acs5.get(("NAME", "B25035_001E", 
                          "B25058_001E", "B25064_001E",
                         "B25077_001E", "B25088_002E"), {'for': 'county:*'})
census_2012_sandy = pd.DataFrame(census_data)

census_2012_sandy = census_2012_sandy.rename(columns={"B25035_001E": "Median House Construction",
                                                      "B25058_001E": "Median Contract Rent",
                                                      "B25064_001E": "Median Gross Rent",
                                                      "B25077_001E": "Median Home Value",
                                                      "B25088_002E": "Median Monthly Owner Cost",
                                                      "NAME": "Name", "county": "county"})
census_2012_sandy.head()
```

```python
# Census API Key
c3 = Census(census_key, year=2015)
```

```python
census_data = c3.acs5.get(("NAME", "B25035_001E", 
                          "B25058_001E", "B25064_001E",
                         "B25077_001E", "B25088_002E"), {'for': 'county:*'})
census_2015_sandy = pd.DataFrame(census_data)

census_2015_sandy = census_2015_sandy.rename(columns={"B25035_001E": "Median House Construction",
                                                      "B25058_001E": "Median Contract Rent",
                                                      "B25064_001E": "Median Gross Rent",
                                                      "B25077_001E": "Median Home Value",
                                                      "B25088_002E": "Median Monthly Owner Cost",
                                                      "NAME": "Name", "county": "county"})
census_2015_sandy.head()
```

```python
merge_one_data = pd.merge(census_2010_sandy, census_2012_sandy, how="right", on="Name")

merge_one_data.head()
```

```python
sandy_complete_housing = pd.merge(merge_one_data, census_2015_sandy, how="right", on="Name")

sandy_complete_housing.head()
```

```python
sandy_reduced = sandy_complete_housing[["Name", "Median House Construction_x", "Median Contract Rent_x", "Median Home Value_x",
                                        "Median House Construction_y", "Median Contract Rent_y", "Median Home Value_y",
                                        "Median House Construction", "Median Contract Rent", "Median Home Value"]]
sandy_reduced.head()                                        
```

```python
sandy_renamed = sandy_reduced.rename(columns={"Median House Construction_x": "Construction_2010",
                                             "Median Contract Rent_x": "Rent_2010",
                                             "Median Home Value_x": "Value_2010",
                                             "Median House Construction_y": "Construction_2012",
                                             "Median Contract Rent_y": "Rent_2012",
                                             "Median Home Value_y": "Value_2012",
                                             "Median House Construction": "Construction_2015",
                                             "Median Contract Rent": "Rent_2015",
                                             "Median Home Value": "Value_2015"})
sandy_renamed.head()
```

```python
sandy_clean = sandy_renamed.loc[sandy_renamed["Name"] == "Atlantic County, New Jersey"]
sandy_clean
```

```python
sandy_housing_df = pd.DataFrame({"Years": [2010, 2012, 2015],
                        "Median Construction": [1973, 1974, 1973],
                        "Median Contract Rent": [828, 881, 893],
                        "Median Home Value": [264400, 248100,225600]})
sandy_housing_df
```

```python
years = ["2010", "2012", "2015"]
construction = [1973, 1974, 1973]
x_axis = np.arange(len(construction))

plt.bar(x_axis, construction, color="b", align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, years)


plt.title("Atlantic City County Median House Construction")
plt.xlabel("Years")
plt.ylabel("Costs")
```
