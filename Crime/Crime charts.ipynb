```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
atlantic_df = pd.read_csv('Atlantic_Crime.csv')

atlantic_df.head()
```

```python
atlantic_chart = atlantic_df[['Year', 'Total']]
atlantic_chart = atlantic_chart.set_index("Year")
atlantic_chart
```

```python
atlantic_chart.plot(kind="bar", color='green')

plt.title("Total crime by year, Atlantic County NJ")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.savefig("Total_Crime_Atlantic.png")
plt.show()
```

```python
atlantic_df = atlantic_df.set_index("Year")
del atlantic_df['Total']
atlantic_line = atlantic_df.transpose()

atlantic_line
```

```python
atlantic_line.loc['Murder', "2010":"2020"].plot(label="Murder")
atlantic_line.loc['Rape', "2010":"2020"].plot(label="Rape")
atlantic_line.loc['Robbery', "2010":"2020"].plot(label="Robbery")
atlantic_line.loc['Assault', "2010":"2020"].plot(label="Assault")
atlantic_line.loc['Burglary', "2010":"2020"].plot(label="Burglary")
atlantic_line.loc['Larceny', "2010":"2020"].plot(label="Larceny")
atlantic_line.loc['Auto Theft', "2010":"2020"].plot(label="Auto Theft")
plt.legend()
plt.title("Crime by type over time, Atlantic County NJ")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.savefig("Crime_Type_Atlantic.png")
plt.show()
```

```python
harris_df = pd.read_csv('Harris_Crime.csv')

harris_df.head()
```

```python
harris_chart = harris_df[['Year', 'Total']]
harris_chart = harris_chart.set_index("Year")
harris_chart
```

```python
harris_chart.plot(kind="bar", color="green")

plt.title("Total crime by year, Harris County TX")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.savefig("Total_Crime_Harris.png")
plt.show()
```

```python
harris_df = harris_df.set_index("Year")
del harris_df['Total']
harris_line = harris_df.transpose()

harris_line
```

```python
harris_line.loc['Murder', "2015":"2019"].plot(label="Murder")
harris_line.loc['Rape', "2015":"2019"].plot(label="Rape")
harris_line.loc['Robbery', "2015":"2019"].plot(label="Robbery")
harris_line.loc['Assault', "2015":"2019"].plot(label="Assault")
harris_line.loc['Burglary', "2015":"2019"].plot(label="Burglary")
harris_line.loc['Larceny', "2015":"2019"].plot(label="Larceny")
harris_line.loc['Auto Theft', "2015":"2019"].plot(label="Auto Theft")
plt.xlim(2015, 2019)
x_ticks = np.arange(2015, 2020, 1)
plt.xticks(x_ticks)
plt.legend()
plt.title("Crime by type over time, Harris County TX")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.savefig("Crime_Type_Harris.png")
plt.show()
```
